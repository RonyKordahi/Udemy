```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
fig = plt.figure(dpi = 200, figsize = (2, 2))

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b);

fig.savefig("testfigure.png", bbox_inches = "tight") # bbox allows us to increase padding to save X and Y axes
```
