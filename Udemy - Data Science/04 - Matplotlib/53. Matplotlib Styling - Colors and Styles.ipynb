```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 11)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

# to see the points from X, pass marker
ax.plot(x, x, color = "crimson", label = "x", ls = "--", marker = "o")

# setting up custom style
lines = ax.plot(x, x + 1, color = "purple", label = "x + 1", lw = 3)
lines[0].set_dashes([5, 2, 1, 2])

# to make markers bigger, use markersize/ms
ax.plot(x, x + 2, color = "blue", marker = "+", lw = 3, ms = 20, label = "x + 2")

# custom markers
ax.plot(x, x - 1, marker = "*", ms = 20, markerfacecolor = "red", markeredgewidth = 2.5, markeredgecolor = "black", label = "x - 1")

ax.legend();
```
