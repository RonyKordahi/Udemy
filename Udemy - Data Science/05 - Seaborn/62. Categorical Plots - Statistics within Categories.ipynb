```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/dm_office_sales.csv")
df.head()
```

```python
# countplot is the graph equivalent of df["division"].value_counts()
# countplot doesn't need a Y axis, only counts what's passed into X

plt.figure(figsize = (10,5), dpi = 200)

sns.countplot(data = df, x = "level of education", hue = "division")

# hue divides it further
```

```python
plt.figure(figsize = (10, 5), dpi = 200)

# estimator is mean by default
# ci (confidence interval) is 95 by default

sns.barplot(data = df, x = "level of education", y = "salary", estimator = np.mean, ci = "sd", hue = "division"); 
```
