```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/StudentsPerformance.csv")
df.head()
```

```python
# more generalized form of other plots, kind parameter displays it according to the one we set
# row/col separates it into a subplot based on categorical value

# catplot calls something called FacetGrid, similar to how pairplot calls PairGrid

sns.catplot(data = df, x = "gender", y = "math score", kind = "box", row = "lunch", col = "test preparation course");
```

```python
# what sns.pairplot() calls

# upper level parameters go here
g = sns.PairGrid(df, hue = "gender")

# mapping
g = g.map_upper(sns.scatterplot) # top mid, top right, mid right
g = g.map_lower(sns.kdeplot) # mid left, bot left, bot mid
g = g.map_diag(sns.kdeplot) # diagonals: top left, mid mid, bot right - SCATTERPLOT USELESS

# legend must be added manually (always at bottom)
g = g.add_legend()
```
