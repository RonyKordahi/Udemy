```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/dm_office_sales.csv")
df
```

```python
# to edit seaborn figure, use matplotlib.pyplot

plt.figure(figsize = (12, 4), dpi = 200)

# increase size of dot basedd on data: size parameter | increase size of all dots: s parameter
sns.scatterplot(x = "salary", y = "sales", data = df, hue = "level of education", style = "level of education", s = 100);

# to save, same code as plt
plt.savefig("test_plot.png")
```
