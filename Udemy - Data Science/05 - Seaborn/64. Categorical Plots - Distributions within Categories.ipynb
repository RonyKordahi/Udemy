```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/StudentsPerformance.csv")
df.head()
```

```python
plt.figure(figsize = (10, 5), dpi = 200)

sns.boxplot(data = df, y = "math score", x = "parental level of education", hue = "test preparation course")

# legend location has to come after plot
plt.legend(bbox_to_anchor = (1, 1));
```

```python
plt.figure(figsize = (10, 5), dpi = 200)

# split parameter splits the violins in half and combines both hues into 1 violin
sns.violinplot(data = df, y = "math score", x = "parental level of education", hue = "test preparation course", split = True)
```

```python
plt.figure(figsize = (10, 5), dpi = 200)

# if too much data, shrink dot size to fit all dots on the screen
# dodge parameter separates the hues
sns.swarmplot(data = df, y = "math score", x = "parental level of education", size = 3,
              hue = "test preparation course", dodge = True);
```

```python
# almost never used
plt.figure(figsize = (10, 5), dpi = 200)

sns.boxenplot(data = df, y = "math score", x = "parental level of education");
```
