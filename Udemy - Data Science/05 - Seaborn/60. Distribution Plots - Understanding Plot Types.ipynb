```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/dm_office_sales.csv")
```

```python
df
```

```python
# can be modified with plt.figure()

sns.rugplot(x = "salary", data = df, height = 0.5)
```

```python
sns.set(style = "darkgrid")
sns.displot(data = df, x = "salary", color = "red", edgecolor = "black", linewidth = 4, kde = True)

# can edit bins with bins
# can edit style with sns.set(style) darkgrid | whitegrid | dark | white | ticks *SETS STYLE FOR THE WHOLE NOTEBOOK*
# all plt customization work for sns too
# can add KDE with kde = True, can add rugplot with rug = True
```

```python
# almost identical to displot, with slightly less 

sns.histplot(data = df, x = "salary")
```

```python
sns.kdeplot(data = df, x = "salary")
```

```python
np.random.seed(42)

sample_ages = np.random.randint(0, 100, 200)
```

```python
sample_ages = pd.DataFrame(sample_ages, columns = ["age"])
```

```python
sns.rugplot(data = sample_ages, x = "age", height = 0.2);
```

```python
# notice how KDE is cut on 0 and 100

sns.displot(data = sample_ages, x = "age", bins = 30, kde = True);
```

```python
# KDE shows probability beyond expected range, add clip to cut it
# to change bandwith, use bw_adjust

sns.kdeplot(data = sample_ages, x = "age", clip = [0, 100], bw_adjust = 0.5);
```
