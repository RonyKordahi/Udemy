```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/StudentsPerformance.csv")
df.head()
```

```python
# kind = "hist" makes them blocks instead, the blocks line up with histograms on the sides
# kind = "kde" makes a black hole (with shade = True)
# kind = "hex" does not have hue

sns.jointplot(data = df, x = "math score", y = "reading score", hue = "gender");
```

```python
# WARNING - CPU/RAM intensive
# grabs all numerical columns and automatically plots them and their relationships to each other
# hue changes the histogram to KDE automatically, histogram can be forced with diag_kind = "hist"
# corner = True erases dupliacates

sns.pairplot(df, hue = "gender");
```
