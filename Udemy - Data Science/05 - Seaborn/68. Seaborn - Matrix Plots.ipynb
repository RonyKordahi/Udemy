```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/country_table.csv")
df
```

```python
# for a heatmap, we need change the index to something with labels

df = df.set_index("Countries")
df
```

```python
# life expectancy is in a different unit than other columns (years vs. rates), we have to remove it
# linewidth (lw) is 0 by default
# annotate (annot) is False by default
# to change colors, use cmap not palette
# center value of color bar can be changed with center parameter

plt.figure(dpi = 200)

sns.heatmap(df.drop("Life expectancy", axis = 1), lw = 0.5, annot = True);
```

```python
# sorts data by hierarchy and clusters them together

sns.clustermap(df.drop("Life expectancy", axis = 1), lw = 0.5, annot = True);
```
